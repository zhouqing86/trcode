```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
a = np.random.normal(size=1000)
bins = np.arange(-4, 5)
histogram = np.histogram(a, bins=bins, normed=True)[0]
bins = 0.5*(bins[1:] + bins[:-1])
b = stats.norm.pdf(bins)
plt.plot(bins, histogram)
plt.plot(bins, b)
plt.show()  
```
